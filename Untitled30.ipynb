Q9. Chart Selection Justification

Explain why different charts were used for different business questions in this analysis.

Sol :
- Bar Charts (Tasks 1 & 6): Chosen because they easily allow viewers to compare distinct categorical values (like product categories) side-by-side.
- Line Charts (Tasks 2 & 5): Ideal for visualizing time-series data, making it easy to identify peaks, valleys, and cumulative historical growth.
- Pie Chart (Task 3): Used effectively here to show a "parts-to-whole" relationship, allowing management to immediately grasp how regions divide the 100% total sales pie.
- Scatter / Bubble Plots (Tasks 4 & 7): Best for analyzing relationships and correlations between multiple continuous numeric variables (e.g., Sales vs. Profit), successfully revealing outlier behaviors like high-revenue/negative-profit orders.
